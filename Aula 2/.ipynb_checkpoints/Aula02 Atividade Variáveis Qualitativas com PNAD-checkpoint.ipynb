```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import os
matplotlib.style.use('ggplot')
```

```python
dados = pd.read_table('pes2013A.txt', sep=" ")
```

```python
dados.head(10)
```

```python
dados.V0302 = dados.V0302.astype('category') #Sexo
dados.V0302.cat.categories = ('Masculino', 'Feminino')

dados.V0404 = dados.V0404.astype('category') #Cor
dados.V0404.cat.categories = ('Indígena', 'Branco', 'Negro', 'Amarelo', 'Pardo', 'Sem declaração')

dados.V0502 = dados.V0502.astype('category') #Natural ou imigrante
dados.V0502.cat.categories = ('Natural', 'Emigrante')

dados.V6002 = dados.V6002.astype('category') #Rede de Ensino
dados.V6002.cat.categories = ('Publica', 'Privada')

dados.V6007 = dados.V6007.astype('category') #Rede de Ensino
dados.V6007.cat.categories = ('Elementar','Ginasial', 'Cietífico','Ens. Fund','Ens. Médio','EJA 1', 'EJA 2', 'Superior', 'Pós Graduação', 'AJA', 'Creche', 'CA', 'Maternal')
```

```python
rondonia = dados[(dados.UF==11)]
acre = dados[(dados.UF ==12)]
amazonas = dados[(dados.UF ==13)]
roraima = dados[(dados.UF ==14)]
para = dados[(dados.UF ==15)]
para = dados[(dados.UF ==15)]
tocantins = dados[(dados.UF ==17)]
```

```python
norte = dados[(dados.UF >= 11) & (dados.UF <= 17)]
print ("Dados Referentes à Região Norte")
norte.head(3)
```

```python
nordeste = dados[(dados.UF >= 21) & (dados.UF <= 29)]
print ('Dados referentes à Região Nordeste')
nordeste.head(3)
```

```python
mg = dados[(dados.UF ==31)]
es = dados[(dados.UF ==32)]
sp = dados[(dados.UF ==35)]
rj = dados[(dados.UF ==33)]
```
