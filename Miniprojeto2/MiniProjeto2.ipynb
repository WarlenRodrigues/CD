### Distribuição Gamma

```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

```python
dados = stats.gamma.rvs(scale=10,loc=4, a = 6, size=100000)
MeanDados = np.mean(dados)
VarDados = np.var(dados)
DPDados = VarDados**(1/2)
```

```python
plt.hist(dados, bins = 20, normed = 1)
plt.axis([0,200,0,0.018])
plt.grid(True)
plt.show()
```

### No gráfico acima temos a representação da população criada a partir da Distribuição Gamma.

```python
n = 2
random = np.random.choice(dados,n)
MeanRand = 0
for i in range(len(random)):
    MeanRand += random[i]
MeanRand = MeanRand/n
```

```python
ListMean = list()

for i in range(100000):
    MeanRand = 0
    random = np.random.choice(dados,n)
    for j in range(len(random)):
        MeanRand += random[j]
    ListMean.append(MeanRand/n)
```

### Inicialmente, definimos que a amostra conteria dois elementos aleatórios da população criada e, em seguida obtivemos a média de 100000 amostras. Como resultado disso temos o seguinte gráfico:

```python
x = DPDados/((n**(1/2)))

t = np.linspace(0,200,4000)
plt.hist(ListMean, bins = 20, normed = True, color='gold')
plt.grid(True)
plt.title("Para 100000 simulações")
plt.ylabel("Densidade")
plt.xlabel("Médias Amostrais")
plt.plot(t,stats.norm.pdf(t,scale=x,loc=MeanDados), color = 'red')
plt.show()
```

### A partir dessa etapa, realizamos simulações alterando o número de elementos dentro de cada amostra, obtendo o gráfico que representa cada uma delas.

```python
#Com quatro elementos dentro da amostra:
n = 4
x = DPDados/((n**(1/2)))

random = np.random.choice(dados,n)
MeanRand = 0
for i in range(len(random)):
    MeanRand += random[i]
MeanRand = MeanRand/n

ListMean = list()

for i in range(10000):
    MeanRand = 0
    random = np.random.choice(dados,n)
    for j in range(len(random)):
        MeanRand += random[j]
    ListMean.append(MeanRand/n)
    
t = np.linspace(0,200,4000)
plt.hist(ListMean, bins = 50, normed = True, color='gold')
plt.grid(True)
plt.title("Para 100000 simulações")
plt.ylabel("Densidade")
plt.xlabel("Médias Amostrais")
plt.plot(t,stats.norm.pdf(t,scale=x,loc=MeanDados), color = 'red')
plt.show()
```

```python
#Com vinte elementos dentro da amostra:
n = 20
x = DPDados/((n**(1/2)))

ListMean = list()

for i in range(10000):
    MeanRand = 0
    random = np.random.choice(dados,n)
    for j in range(len(random)):
        MeanRand += random[j]
    ListMean.append(MeanRand/n)
    
t = np.linspace(0,200,4000)
plt.hist(ListMean, bins = 20, normed = True, color='gold')
plt.grid(True)
plt.title("Para 100000 simulações")
plt.ylabel("Densidade")
plt.xlabel("Médias Amostrais")
plt.plot(t,stats.norm.pdf(t,scale=x,loc=MeanDados), color = 'red')
plt.show()
```

```python
#Com cinquenta elementos dentro da amostra:
n = 50
x = DPDados/((n**(1/2)))

listaX1 = list()
listaX9 = list()

random = np.random.choice(dados,n)
MeanRand = 0
for i in range(len(random)):
    MeanRand += random[i]
MeanRand = MeanRand/n

ListMean = list()

for i in range(100000):
    MeanRand = 0
    random = np.random.choice(dados,n)
    for j in range(len(random)):
        MeanRand += random[j]
        
    listaX1.append(random[0])
    listaX9.append(random[8])
    
    ListMean.append(MeanRand/n)
    
t = np.linspace(0,200,4000)
plt.hist(ListMean, bins = 20, normed = True, color='gold')
plt.grid(True)
plt.title("Para 100000 simulações")
plt.ylabel("Densidade")
plt.xlabel("Médias Amostrais")
plt.plot(t,stats.norm.pdf(t,scale=x,loc=MeanDados), color = 'red')
plt.show()
```

### A variância das amostras é definida pela variância das populações dividida pelo número de elementos nas amostras. 

### Pelos gráficos, percebemos que: quanto maior o número de elementos nas amostras analisadas, mais estreita é a curva obtida, o que significa que a variância é menor, aproximando-se de uma distribuição Normal ou Gaussiana.

```python
plt.hist(listaX1, bins=20,normed = True, color="lightgreen")
plt.plot(t,stats.gamma.pdf(t,scale=10,loc=4, a = 6), color = 'red')
plt.axis([0,200,0,0.018])
plt.grid(True)
plt.title("X1")
plt.xlabel("Medidas Amostrais")
plt.ylabel("Densidade")
plt.show()
```

```python
plt.hist(listaX9, bins=20,normed = True, color="lightgreen")
plt.plot(t,stats.gamma.pdf(t,scale=10,loc=4, a = 6), color = 'red')
plt.axis([0,200,0,0.018])
plt.grid(True)
plt.title("X9")
plt.xlabel("Medidas Amostrais")
plt.ylabel("Densidade")
plt.show()
```

```python
CorreX1X9 = np.corrcoef(listaX1,listaX9)
print("Correção X1 e X9: {0}".format(CorreX1X9[0][1]))
```

### Observa-se que a distribuição do primeiro e do nono elemento de cada uma das 100000 segue o modelo de uma distribuição Gamma. Além disso, podemos observar que é a mesma curva é exatamente a mesma curva referente a população. Isso se deve a Correção X1 (primeiro termo de cada amostra) e X9 (nono termo de cada amostra) é aproximadamente 0, assim como calculado.
